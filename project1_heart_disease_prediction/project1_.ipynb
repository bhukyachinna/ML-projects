```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
heart_data = pd.read_csv('framingham.csv')
```

```python
heart_data.head()
```

```python
heart_data.isnull().sum()
```

```python
for col in heart_data.columns:
    if heart_data[col].isnull().sum() > 0:
        heart_data[col].fillna(heart_data[col].mean(), inplace = True)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
y = heart_data['TenYearCHD']
x = heart_data.drop('TenYearCHD', axis=1)
x
```

```python
std = MinMaxScaler()
std.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=2, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logi = LogisticRegression(max_iter=2500)
logi.fit(x_train, y_train)
```

```python
pred = logi.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred) * 100
```
