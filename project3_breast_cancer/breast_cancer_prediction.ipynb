```python
import pandas as pd
```

```python
bc_df = pd.read_csv('data.csv')
```

```python
bc_df.head()
```

```python
bc_df.shape
```

```python
bc_df.isnull().sum()
```

```python
bc_df.dropna(axis=1, inplace=True)
```

```python
bc_df.drop(['id'], axis=1, inplace=True)
```

```python
bc_df
```

```python
bc_df['diagnosis'].value_counts()
```

```python
bc_df['diagnosis'].replace(to_replace={"B": 0, "M": 1}, inplace=True)
```

```python
y = bc_df['diagnosis']
```

```python
x = bc_df.drop(['diagnosis'], axis=1)
x
```

```python
from sklearn.preprocessing import MinMaxScaler
std = MinMaxScaler()
std.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.20, random_state=1)
```

```python
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(x_train, y_train)
```

```python
pred = lin.predict(x_test)
```

```python
prediction = list()
for i in pred:
    if i >= 0.5:
        prediction.append(1)
    else:
        prediction.append(0)
```

```python
from sklearn.metrics import accuracy_score
print(f'testing : {accuracy_score(y_test, prediction) * 100}')
```

```python
pred_train = lin.predict(x_train)
```

```python
prediction_train = list()
for i in pred_train:
    if i >= 0.5:
        prediction_train.append(1)
    else:
        prediction_train.append(0)
```

```python
print(f'training : {accuracy_score(y_train, prediction_train) * 100}')
```
